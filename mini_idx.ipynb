```python
%matplotlib inline
from new_methods.new_methods import My_NuScenes
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
import os.path as osp
import pythreejs as pjs
import pickle

# load instance
nusc = My_NuScenes(version='v1.0-trainval', dataroot='/data/datasets/nuscenes', verbose=True)
nusc_mini = My_NuScenes(version='v1.0-mini', dataroot='/data/datasets/nuscenes', verbose=True)

# load pickle files
with open('/data/datasets/nuscenes/nuscenes_infos_train.pkl', 'rb') as f:
	train_infos = pickle.load(f)
with open('/data/datasets/nuscenes/nuscenes_infos_val.pkl', 'rb') as f:
	val_infos = pickle.load(f)

print(len(train_infos), len(val_infos))
print(len(nusc_mini.sample), len(nusc.sample))
```

```python
# make token list
sample_token_mini = [nusc_mini.sample[i]['token'] for i in range(len(nusc_mini.sample))]
sample_token_train = [train_infos[i]['sample_token'] for i in range(len(train_infos))]
sample_token_val = [val_infos[i]['sample_token'] for i in range(len(val_infos))]
sample_token_trainval = [nusc.sample[i]['token'] for i in range(len(nusc.sample))]
print(len(sample_token_mini), len(sample_token_train), len(sample_token_val), len(sample_token_trainval))
```

```python
# initialize idx
keys = ['trainorval', 'scene', 'mini_idx', 'train_idx', 'val_idx', 'trainval_idx', 'sample_token']
idx_dict = {
    keys[0]: [],
    keys[1]: [],
    keys[2]: [],
    keys[3]: [],
    keys[4]: [],
    keys[5]: [],
    keys[6]: [],
}

# make scene name list
scene_names_mini_list = [nusc.get('scene', nusc_mini.sample[i]['scene_token'])['name'] for i in range(len(nusc_mini.sample))]

# calculate scene, idx
for i in range(len(sample_token_mini)):
    scene_name = scene_names_mini_list[i]
    token = sample_token_mini[i]
    mini_idx = i
    idx_dict['scene'].append(scene_name)
    idx_dict['mini_idx'].append(str(mini_idx))
    idx_dict['sample_token'].append(token)
    try:
        train_idx = sample_token_train.index(token)
        idx_dict['trainorval'].append('train')
        idx_dict['train_idx'].append(train_idx)
    except ValueError:
        idx_dict['train_idx'].append(None)
    try:
        val_idx = sample_token_val.index(token)
        idx_dict['trainorval'].append('val')
        idx_dict['val_idx'].append(val_idx)
    except ValueError:
        idx_dict['val_idx'].append(None)
    try:
        trainval_idx = sample_token_trainval.index(token)
        idx_dict['trainval_idx'].append(trainval_idx)
    except ValueError:
        print(f"This token is not valid: {token}")

# check all values availabe (404, 404, 404, 404)
print(len(idx_dict['trainorval']), len(idx_dict['scene']), len(idx_dict['mini_idx']), len(idx_dict['train_idx']), len(idx_dict['val_idx']), len(idx_dict['trainval_idx']), len(idx_dict['sample_token']))
```
